```python
%matplotlib inline
from os import listdir
from os.path import isfile, join
import numpy as np
from glob import glob
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib
import statsmodels.stats.multitest as smt
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
from adjustText import adjust_text
from scipy import stats
import itertools
import matplotlib.patches as mpatches
```

```python
## Read summary file with total number of variants by effect class
dir_path = '/Users/osipova/Documents/LabDocs/Leatherback_popgen/'
file_name = 'all_samples.total_variant_by_effect.tsv'
file_data = glob(dir_path + file_name)[0]
df = pd.read_csv(file_data, sep='\t')


## Read samples location info
dir_path = '/Users/osipova/Documents/LabDocs/Leatherback_popgen/'
file_name = 'dc_ALL_rounds_info.tsv'
file_data = glob(dir_path + file_name)[0]
INFO = pd.read_csv(file_data, sep='\t')[['Sample_ID', 'country']]
INFO.rename(columns={'Sample_ID':'sample', 'country':'country'}, inplace=True)


## Assign location to each sample
df_all = pd.merge(df, INFO, on='sample')
df_all.head(3)
```

## Plot total number of variants by effect

```python
effects = list(set(df_all.columns) - set(['sample', 'country']))

fig, axes = plt.subplots(2, 2, figsize = (11, 11))

for i in range(len(effects)):
    j = i % 2
    k = i // 2
    axi = axes[j][k]
    effect = effects[i]
    sns.boxplot(ax=axi, data=df_all, x='country', y=effect, showfliers=False)
    sns.stripplot(ax=axi, data=df_all, x='country', y=effect, color='black', alpha=0.6, s=4)
    plt.setp(axi.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    axi.set_ylabel('total number of {} impact variants'.format(effect))
    axi.set_xlabel('')
    
    
sns.despine(fig=None, ax=None, left=False, bottom=False, offset=None, trim=False)
fig.tight_layout()

pdf_name = 'total_number_variants_by_effect.pdf'
# fig.savefig(pdf_name, bbox_inches='tight')
```

## Plot fraction of variants by effect

```python
df_all['total'] = df_all['HIGH'] + df_all['LOW'] + df_all['MODERATE'] + df_all['MODIFIER']

for effect in effects:
    effect_fraction = '{}_f'.format(effect)
    df_all[effect_fraction] = df_all[effect] / df_all['total'] * 100
df_all.head(3)
```

```python
fig, axes = plt.subplots(2, 2, figsize = (11, 11))

for i in range(len(effects)):
    j = i % 2
    k = i // 2
    axi = axes[j][k]
    effect = effects[i]
    sns.boxplot(ax=axi, data=df_all, x='country', y='{}_f'.format(effect), showfliers=False)
    sns.stripplot(ax=axi, data=df_all, x='country', y='{}_f'.format(effect), color='black', alpha=0.6, s=4)
    plt.setp(axi.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    axi.set_ylabel('% of {} impact variants'.format(effect))
    axi.set_xlabel('')
    
    
sns.despine(fig=None, ax=None, left=False, bottom=False, offset=None, trim=False)
fig.tight_layout()

pdf_name = 'percent_variants_by_effect.pdf'
# fig.savefig(pdf_name, bbox_inches='tight')
```
