## 2.1. Plot sliding window for one sample

```python
def make_position(dfi, col_chr, col_start):
    ## sorts by chromosome number and adds absolute postion column
    
    exclude = ['chrZ', 'chrW', 'chrNA', 'chrMT']
    dfi['COUNT'] = dfi[col_chr].str.replace('scaffold_', '')
    dfi['COUNT'] = dfi[col_chr].str.replace('SUPER_', '')
    
    dfi_num = dfi[ ~dfi[col_chr].isin(exclude)]
    dfi_Z = dfi[dfi[col_chr] == 'chrZ']
    dfi_W = dfi[dfi[col_chr] == 'chrW']
    
    dfi_num['COUNT'] = pd.to_numeric(dfi_num['COUNT'])
    dfi_num = dfi_num.sort_values(by=['COUNT', col_start])

    dfi_merged = pd.concat([dfi_num, dfi_Z, dfi_W], ignore_index=True)
    dfi_merged['POSITION'] = [x / 10 for x in range(dfi_merged.shape[0])]
    return dfi_merged
```

```python
warnings.filterwarnings('ignore')

samples = ['dc_100028', 'dc_37164', '23560', '88837', '96545']

fig, axes = plt.subplots(len(samples), 1, figsize=(20, 2.7*len(samples)))

for i in range(len(samples)):
    sample = samples[i]
    axi = axes[i]
    dfi = df_all[df_all["sample"] == sample]
#     dfi = dfi[dfi['chrom'].isin(['SUPER_1', 'SUPER_6', 'SUPER_7'])]
#     dfi = dfi[dfi['chrom'].isin(['SUPER_6'])]
   
    df_merged = make_position(dfi, 'chrom', 'window_start')

    N = len(set(df_merged['COUNT']))
    p = (['black', 'grey'] * (int(N / 2) + 1))[:N]

    sns.scatterplot(ax=axi, data=df_merged, x='POSITION', y='hets', hue='chrom', 
                    s=10, linewidth=0, alpha=0.9, palette=p)

    # Get midpoints for each chromosome
    chr_mids = df_merged.groupby('chrom')['POSITION'].agg(['min', 'max'])
    chr_mids['mid'] = (chr_mids['min'] + chr_mids['max']) / 2

    # Add labels at midpoints
    for chrom, row in chr_mids.iterrows():
        axi.text(row['mid'], axi.get_ylim()[1]*0.95, chrom, 
                 ha='center', va='top', fontsize=8, color='black')

    sns.despine(ax=axi)
    axi.legend([], [], frameon=False)
    rookery = list(set(dfi['rookery']))[0]
    axi.set_title(f'{sample} {rookery}')
    axi.set_xlabel('POSITION, Mb')

fig.tight_layout()
```

# 5. Temporal

## 5.1. heterozygosity = f(t) Mexico and Virgin Islands

```python
group = 'rookery'
group_col = 'MU'
x = 'year'
y = 'f_hets'
p = color_dict[group_col]

# temporal_groups = ['Mexico', 'Virgin Islands', 'Indonesia']
temporal_groups = ['Mexico', 'Virgin Islands']
df = df_all[df_all[group].isin(temporal_groups)].dropna(subset=[x])
df['stage'] = df['stage'].fillna('adult')
df['outlier'] = df['outlier'].fillna('include')
df = df[df['outlier'] == 'include']

fig, axes = plt.subplots(figsize=(7, 4))

if x == 'year':
    style = 'stage'
    markers_dict = {'adult': 'o', 'hatchling': 'v'}
else:
    style = None
    markers_dict = None

sns.scatterplot(
    data=df, x=x, y=y, hue=group_col,
    alpha=0.8, edgecolor='k', linewidth=0.2,
    style=style, markers=markers_dict,
    palette=p, ax=axes)

axes.set_ylabel('heterozygosity')
sns.move_legend(axes, "upper left", bbox_to_anchor=(1, 1))


# GLM
stats_results = []

for mu, subdf in df.groupby(group_col):

    # Fit GLM
    formula = f"{y} ~ {x} + stage"
    model = smf.glm(formula=formula, data=subdf).fit()
    
    # Fit GLM
    eps = 1e-6
    tmp = subdf.copy()
    clipped = tmp[y].clip(eps, 1 - eps)
    tmp['_y'] = np.log(clipped / (1 - clipped))  # logit
    model = smf.glm(f"_y ~ {x} + C(stage)", data=tmp).fit()


    slope = model.params[x]
    p_value = model.pvalues[x]

    stats_results.append({"MU": mu, "slope": slope, "pvalue": p_value})

    color = p[mu] if isinstance(p, dict) else None

    p_thresh = 1
    if p_value < p_thresh:

        xvals = np.linspace(subdf[x].min(), subdf[x].max(), 100)
        pred_df = pd.DataFrame({x: xvals})

        if 'stage' in subdf.columns:
            pred_df['stage'] = subdf['stage'].mode()[0]

        pred = model.get_prediction(pred_df)
        pred_mean = expit(pred.predicted_mean)
        ci = pred.conf_int()
        lower = expit(ci[:, 0])
        upper = expit(ci[:, 1])

        # Regression line
        axes.plot(xvals, pred_mean, color=color, lw=2, alpha=0.9)

        # Confidence interval shading
        axes.fill_between(xvals, lower, upper, color=color, alpha=0.06)

        # p-value annotation
        midx = np.median(xvals)
        maxy = subdf[y].max()
        axes.text(midx, maxy, f"p = {p_value:.2e}", ha="center", va="bottom", fontsize=12)


# Print stats
results_df = pd.DataFrame(stats_results)
print(results_df)

sns.despine(fig=None, ax=None, left=False, bottom=False)
fig.tight_layout()

pdf_name = 'glm.temporal.heterozygosity.year.pdf'
# fig.savefig(pdf_name, bbox_inches='tight')
```
